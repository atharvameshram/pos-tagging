```python
vocabulary = dict()
idx = 0

for row in df.values:
  if row[1] in vocabulary:
    vocabulary[row[1]]['occurrences'] += 1
  else:
    vocabulary[row[1]] = dict()
    vocabulary[row[1]]['index'] = idx
    vocabulary[row[1]]['occurrences'] = 1
    idx += 1

print(vocabulary)
print(len(vocabulary))
```

```python
threshold = 3

vocabulary['<unk>'] = dict()
vocabulary['<unk>']['occurrences'] = 0
vocabulary['<unk>']['index'] = -1
rare_words_to_remove = list()

for word in vocabulary:
  if vocabulary[word]['occurrences'] < 3:
    vocabulary['<unk>']['occurrences'] += vocabulary[word]['occurrences']
    rare_words_to_remove.append(word)

for word in rare_words_to_remove:
  vocabulary.pop(word)

print(vocabulary)
print(len(vocabulary))
```

```python
  #Remove rare words from dataframe
mask = df['word_type'].isin(rare_words_to_remove)
df = df[~mask]
len(df['word_type'].value_counts())
```

Q3 Greeedy Decoding

```python
dev_data = pd.read_table('dev', names=['index', 'word_type', 'pos'])
dev_data.head()
```

```python
em_col_list = list(em_column_names.keys())
```

```python
def predictGreedyPOS(prev_pos, curr):
  argmax, pos = 0, ''

  try:
    key = em_row_names[curr]
  except KeyError:
    key = em_row_names['Pierre']

  for idx, e in enumerate(emission.iloc[key, 1:]):
    if e == 0:
      continue

    try:
      row_val = trans_r_names[prev_pos]
    except KeyError:
      row_val = trans_r_names['NNP']

    temp = e * transmission.iloc[row_val, trans_col_names[em_col_list[idx+1]]]
    if temp > argmax:
      argmax = temp
      pos = em_col_list[idx+1]

  return pos
```

```python
prev_pos = '<BOS>'
with open("dev_predict", "w") as file:
  for idx, row in enumerate(dev_data.values):
    prev_pos = predictGreedyPOS(prev_pos, row[1])
    file.write(f'{row[0]}\t{row[1]}\t{prev_pos}\n')

    if(idx+1 != len(dev_data.values) and dev_data.values[idx+1][0] == 1):
      file.write('\n')
      prev_pos = '<BOS>'
```

```python
def eval(pred, gold):
  total = 0
  corr = 0
  try:
      with open(gold, 'r') as gf, open(pred, 'r') as pf, open('outputs.txt', 'w') as offile:
          pf_count = 0
          pf_lines = pf.readlines()
          for gline in gf.readlines():
              gline = gline.strip()
              if len(gline) == 0:

                      # handles missing blank line issue
                      if len(pf_lines[pf_count].strip()) == 0:
                          pf_count+=1
                      continue
              tokens = gline.split()

              # print tokens
              gidx = tokens[0]
              gword = tokens[1]
              glabel = tokens[2]

              pline = pf_lines[pf_count]
              pline = pline.strip()
              tokens = pline.split()

              pword = tokens[1] if len(tokens) > 1 else ''    # handles blank line issue if wrong dev dataset is used
              plabel = tokens[2] if len(tokens) > 2 else ''   # handles missing label issue

              #if gidx != pidx:
              #    print('warning: index mismatch: {}, {}'.format(gidx, pidx))

              #if gword != pword:
              #    print('warning: word mismatch: {}, {}'.format(gword, pword))

              total += 1
              if glabel == plabel:
                  corr += 1

              pf_count += 1

              offile.write(' '.join([gword, pword, glabel, plabel, '\n']))

  except:
      print(repr(gline), repr(pline), total)

  print("total: {}, correct: {}, accuracy: {:.2f}%".format(total, corr, float(corr) * 100 / total))
```

```python
eval('dev_predict', 'dev')
```

Viterbi Decoding Algorithm

```python
sentenceIndex = list(dev_data[dev_data['index'] == 1].index)

print(sentenceIndex)
```

```python
em_col_list_updated = em_col_list[1:]
```

```python
k = len(em_col_list_updated)
def predictViterbiPOS(sentence):
  m = len(sentence)
  pos = [''] * m
  piMatrix = np.zeros((m, k))

  try:
    key = em_row_names[sentence[0]]
  except KeyError:
    key = em_row_names['Pierre']

  for i in range(0, k):
    e = emission.iloc[key, i+1]
    t = transmission.iloc[trans_r_names['<BOS>'], trans_col_names[em_col_list_updated[i]]]
    temp = t * e

    if temp > piMatrix[0][i]:
      piMatrix[0][i] = temp
      pos[0] = em_col_list_updated[i]


  for j in range(1, m):
    try:
      key = em_row_names[sentence[j]]
    except KeyError:
      key = em_row_names['Pierre']

    for i in range(0, k):
      for x in range(0, k):
        row_val = trans_r_names[em_col_list_updated[x]]

        e = emission.iloc[key, i+1]
        t = transmission.iloc[row_val, trans_col_names[em_col_list_updated[i]]]
        temp = piMatrix[j-1][x] * t * e

        if temp > piMatrix[j][i]:
          piMatrix[j][i] = temp
          pos[j] = em_col_list_updated[i]

  return pos
```

```python
with open("dev_viterbi_predict", "w") as file:
  for idx in range(0, len(sentenceIndex)):
    if(idx+1 == len(sentenceIndex)):
      sentence = dev_data.iloc[sentenceIndex[idx]:, 1]
    else:
      sentence = dev_data.iloc[sentenceIndex[idx]:sentenceIndex[idx+1], 1]

    pos = predictViterbiPOS(sentence.tolist())

    for i, word in enumerate(sentence):
      file.write(f'{i+1}\t{word}\t{pos[i]}\n')
    file.write('\n')
```

```python
eval('dev_viterbi_predict', 'dev')
```
